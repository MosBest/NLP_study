```python
from urllib.request import urlopen
import re
from bs4 import BeautifulSoup
```

```python
url = "https://baike.baidu.com/item/%E5%8C%97%E4%BA%AC%E5%9C%B0%E9%93%81/408485"
html = urlopen(url).read().decode("utf-8")
soup = BeautifulSoup(html, features='html.parser')
all_href = soup.find_all('a')
```

```python
s = [re.findall("^<a href=(\"/item/.*\") .*\">北京地铁(\d\d?).*线", str(string)) for string in all_href]
s = [i for i in s if i]
```

```python
s
```

```python
data = {}
for i in s:
    key = i[0][1]
    if key not in data.keys():
        data[key] = 'https://baike.baidu.com' + i[0][0][1:-1]
```

```python
data
```

```python
all_route = []
for key, value in data.items():
    urls = value
    html = urlopen(urls).read().decode("utf-8")
    ss = re.findall(">(.{1,5}——.{1,5})</th><td width=\"[0-9]{1,5}\" align=\"\w+\" valign=\"\w+\">([0-9]{1,10})米?</td>", html)
    if len(ss):
        all_route.append((key, ss))
    
```

```python
station = {}
for i in all_route:
    key, ss = i
    for j in ss:
        route, distance = j
        start, end = route.split("——")
        if start not in station.keys():
            station[start] = [(end, distance, key)]
        else:
            station[start].append( (end, distance, key))
        start, end = end, start
        if start not in station.keys():
            station[start] = [(end, distance, key)]
        else:
            station[start].append( (end, distance, key))        
```

```python
station_copy = station.copy()
```

```python
station_copy
```

```python
def all_route(start, end, station_copy, policy = 1, by_way=None):
    if policy == 1:
        # 路程最短优先
        Shortest_Path_Priority(start, end, station_copy, by_way)
    else:
        # 最少换乘优先
        Minimum_Transfer_Priority(start, end, station_copy, by_way)
```

```python
def Shortest_Path_Priority(start, end, station_copy, by_way = None):
    way = [[(start, '0', '0')]]
    min_length = 9999999 
    reachable_path = []

    while len(way)!=0:
        new_route = way.pop(0)
        next_station = new_route[-1]
        
        for i in station_copy[next_station[0]]:

                
            if i_in_way(i, new_route):
                continue
                
            if i[0] == end:
                finally_route = new_route + [i]
                length = cal_path(finally_route)
                if min_length > length:
                    name = []
                    for h in finally_route:
                        name.append(h[0])
                    flagg = True
                    for h in by_way:
                        if h not in name:
                            flagg = False
                            break
                    if flagg:
                        min_path = finally_route
                        min_length = length
                
            else:
                finally_route = new_route + [i]
                length = cal_path(finally_route)
                if min_length < length:
                    continue
                else:               
                    way.append(finally_route)
                
    print(min_path)
        
                
            
```

```python
def i_in_way(i, way):
    for ways in way:
        if i[0] == ways[0]:
            return True
    return False
```

```python
def cal_path(finally_route):
    sum_ = 0
    for i in finally_route:
        sum_ += int(i[1])
    return sum_
```

```python
def cal_transfer(finally_route):
    a = []
    for i in finally_route:
        a.append(i[2])
    a = set(a)
    return len(a)
```
