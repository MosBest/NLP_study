2. 什么是数据驱动？数据驱动在这个程序里如何体现？
    2.1：
        根据第一节课的理解：就是企业在生成过程中，不希望程序的内容随着新情况，新数据的改变而改变。程序设计者 希望自己的程序可以自己处理新数据，新情况。
        以下为百度：
        数据驱动是通过移动互联网或者其他的相关软件为手段采集海量的数据，将数据进行组织形成信息，之后对相关的信息进行整合和提炼，在数据的基础上经过训练和拟合形成自动化的决策模型。当新的情况发生，新数据输入的时候，系统可以用前面建立的模型以人工智能的方式直接进行决策。
    2.2：
        在本次程序训练中，我们设计"?*X"，它可以匹配不同长度，不同情况的数据,比"?X"考虑的范围大很多。只是这种处理回答是可能会风马牛不相及。但是对话过程中什么是“回答对了”,“什么是错的”。本就是lp的一个难点。
    
3. 数据驱动与 AI 的关系是什么？
        根据上面理解，数据驱动就是根据以往的信息，训练一个模型，这个模型可以处理以后新的情况，新问题。这就像人的成长一样，根据以往的经验，学到的内容，从而处理以后的问题。
        人用的是大脑，机器用的则是“数据驱动”。
        所以，我认为 强大的硬件设备 + 数据驱动 == AI
