```python
在一个大量文本下，每个词出现的次数是
N  N/2 N/3 ... ... 每个词出现的次数下降的非常快
即是长尾问题，约到后面，数据越少，越有可能不具备统计学的意义
可对其取log，解决长尾问题
```

```python
frequences_all = [f for w, f in words_count.most_common()]
```

```python
frequences_sum = sum(frequences_all)
```

```python
def get_prob(word):
    esp = 1/ frequences_sum
    if word in words_count:
        return  words_count[word] / frequences_sum
    else:
        return esp
```

```python
get_prob('我们')
```

```python
def product(numbers):
    return reduce(lambda n1, n2:n1 * n2 , numbers)
```

```python
product([1,2,3,4,5,6,9])
```

```python
# one gram
def language_model_one_gram(string):
    words = cut(string)
    return product([get_prob(w) for w in words])
```

```python
language_model_one_gram("长征火箭下周发射")
```

```python
sentences = '''
这是一个比较正常的句子
这个一个比较稀罕的句子
小明毕业于清华大学
小明毕业于秦华大学
'''.split()
```

```python
for s in sentences:
    print(s, language_model_one_gram(s))
```

```python
need_compared = [
    "今天晚上请你吃大餐,我们一起吃日料, 明天晚上请你吃大餐,我们一起吃苹果",
    "真事一只好看的小猫 真是一只好看的小猫",
    "我去吃火锅，今晚 今晚我去吃火锅"
]

for s in need_compared:
    s1, s2 = s.split()
    p1, p2 = language_model_one_gram(s1), language_model_one_gram(s2)
    
    better = s1 if p1 > p2 else s2
    
    print('{} is more posssible'.format(better))
    print('-'*4 + '{} with probility {}'.format(s1, p1))
```

```python
1-gram 认为 每个词之间是独立的 即 p(x1|x2) = p (x1)
则1-gram 就是P(s) = P(w1 , w2, w3, w4, w5,......) = P(x1) * P(x2) * P(x3) * P(x4) * P(x5)*......
```

# 2-Gram

```python
模型就是 建立一个 东西（比如数学公式），来表达你要处理的问题
```

```python
2-gram 是 P(s) = P(w1 , w2, w3, w4, w5,......) = P(x1) * P(x2|w1) * P(x3|w2) * P(x4|w3) * P(x5|w4) * P(x6|w5) * ......
```
